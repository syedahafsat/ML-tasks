```python
import seaborn as sns
import pandas as pd

# Load the data
iris = sns.load_dataset('iris')

# Look at the first few rows
print(iris.head())
```

```python
# Check for missing values
print(iris.isnull().sum())

# Check for duplicates
print("Duplicates:", iris.duplicated().sum())

# Remove duplicates
iris = iris.drop_duplicates()
```

```python
iris.info()
```

```python
print(iris.describe())

# This shows averages, like sepal length is about 5.8 cm. Next, letâ€™s compare the flower types:

print(iris.groupby('species').mean())
```

```python
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=iris, x=column, hue='species')
    plt.title(f'{column}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f'{column}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Width')
plt.show()
```
